```python
from biggr_maps import map
import json
import math
from biggr_maps.escher import Builder, display_builder
```

```python
m = map.Map(name="test", description="Test map", canvas=(-1000, -1000, 2000, 2000))
```

```python
n1_info = dict(
    bigg_id="mal__L_c",
    name="L-Malate",
    node_is_primary=True,
    x=0,
    y=500,
)
n2_info = dict(
    bigg_id="oaa_c",
    name="Oxaloacetate",
    node_is_primary=True,
    x=0,
    y=-500,
)
```

```python
n3_info = dict(
    bigg_id="nad_c",
    name="Nicotinamide adenine dinucleotide",
    node_is_primary=False,
)
n4_info = dict(
    bigg_id="nadh_c",
    name="Nicotinamide adenine dinucleotide - reduced",
    node_is_primary=False,
)
n5_info = dict(
    bigg_id="h_c",
    name="H+",
    node_is_primary=False,
)
```

```python
mid = map.MidMarkerNode(x=-300, y=0)
```

```python
reaction = map.AutoReaction(
    angle=-0.5 * math.pi,
    unit=60,
    name="Malate dehydrogenase",
    bigg_id="MDH",
    mid_marker=mid,
)
```

```python
reaction.add_metabolite(n1_info, -1, b1_scale=0.6)
reaction.add_metabolite(n2_info, 1, b1_scale=0.6)
reaction.add_metabolite(n3_info, -1)
reaction.add_metabolite(n4_info, 1)
reaction.add_metabolite(n5_info, 1)
```

```python
m.add_reaction(reaction)
```

```python
with open("../data/test.json", "w") as f:
    json.dump(m.to_escher(), fp=f, indent=4)
```

```python
builder = Builder(map_json="../data/test.json")
iframe = display_builder(builder)
```
