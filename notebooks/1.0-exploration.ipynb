```python
from biggr_maps import map
import json
import math
from biggr_maps.escher import Builder, display_builder
```

```python
m = map.Map(name="test", description="Test map", canvas=(-200, -200, 2000, 2000))
```

```python
n1_info = dict(
    bigg_id="mal__L_c",
    name="L-Malate",
    node_is_primary=True,
)
n1 = map.MetaboliteNode(
    x=6.858, y=627.7302, label_x=33.7652265625, label_y=621.384496875, **n1_info
)
```

```python
n2_info = dict(
    bigg_id="oaa_c",
    name="Oxaloacetate",
    node_is_primary=True,
)
n2 = map.MetaboliteNode(
    x=78.3413984375,
    y=1166.88234375,
    label_x=78.3413984375,
    label_y=1166.88234375,
    **n2_info
)
```

```python
n3_info = dict(
    bigg_id="nad_c",
    name="Nicotinamide adenine dinucleotide",
    node_is_primary=False,
)
n3 = map.MetaboliteNode(
    x=48.1480390625,
    y=786.964575,
    label_x=25.973234375,
    label_y=827.157934375,
    **n3_info
)
```

```python
n4_info = dict(
    bigg_id="nadh_c",
    name="Nicotinamide adenine dinucleotide - reduced",
    node_is_primary=False,
)
n5_info = dict(
    bigg_id="h_c",
    name="H+",
    node_is_primary=False,
)
```

```python
mid = map.MidMarkerNode(x=-3.8334062499998254, y=900.7116453125)
multi1 = map.MultiMarkerNode(x=-3.8334062499998254, y=852.079809375)
multi2 = map.MultiMarkerNode(x=-3.2757890624998254, y=932.1315671875)
```

```python
reaction = map.AutoReaction(
    angle=1.3*math.pi,
    unit=60,
    name="Malate dehydrogenase",
    bigg_id="MDH",
    mid_marker=mid,
)
```

```python
reaction.add_metabolite(n1_info, -1)
reaction.add_metabolite(n2_info, 1)
reaction.add_metabolite(n3_info, -1)
reaction.add_metabolite(n4_info, 1)
reaction.add_metabolite(n5_info, 1)
```

```python
m.add_reaction(reaction)
```
